# Spontaneous vs. Stimulated Emissions in Lasers

## Description

&emsp;&nbsp;When powering a laser, there is a direct correlation to the power being put in and the power of the lasers beam that is being output. There are three main stages associated with the output power of the laser. These include: below threshold, at threshold, and above threshold. When the input power is below threshold, the laser is just a dull and dimly lit light source that acts like an LED rather than a laser. At threshold, is when the laser begins to lase and its output light starts to directionalize itself into an actual laser beam rather than an LED source. Then, when the input power is above threshold, the laser is fully lasing and all of its output light is very directional focused into a single beam. This chnage in the lasers beam occurs because of Spontneous and Stimulated Emissions within the laser itself. A spontaneous emission occurs when an atom is exiting an excited state and releases a photon randomly. This random photon has a random direction and wavelength. This means that the color of the photon is random and the light emitted is spread out. On the other hand, stimulated emission is when an atom is already in an excited state and a photon hits the atom and causes it to drop to a lower energy level and emit a second photon. However, this second photon is emitted in the exact same direction as the incident photon and is also the exact same wavelength as the incident photon. This is what happens when the laser begins to reach threshold. At threshold, the laser begins stimulated emission rather than spontaneous emission and begins lasing.
