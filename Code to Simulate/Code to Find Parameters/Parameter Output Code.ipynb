```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def gaussian(x, A, x0, sigma, offset):
    return A * np.exp(-(x - x0) ** 2 / (2 * sigma**2)) + offset

def load_data(filename):
    data = []
    with open(filename, 'r') as file:
        for line in file:
            if line.strip and line[0].isdigit():
                data.append(list(map(float, line.split())))
    data = np.array(data)
    return data[:, 0], data[:, 1]

def find_parameters(filename): 
    wavelength, intensity = load_data(filename)
    
    A_guess = np.max(intensity)
    x0_guess = wavelength[np.argmax(intensity)]
    sigma_guess = 10
    offset_guess = np.min(intensity)
    
    p0 = [A_guess, x0_guess, sigma_guess, offset_guess]
    
    params, covariance = curve_fit(gaussian, wavelength, intensity, p0=p0, maxfev=20000)
    return params, covariance

files= [
        "Data For Curve Fit/2mA_clean.txt", "Data For Curve Fit/4mA_clean.txt", "Data For Curve Fit/6mA_clean.txt", 
        "Data For Curve Fit/8mA_clean.txt", "Data For Curve Fit/10mA_clean.txt", "Data For Curve Fit/12mA_clean.txt", 
        "Data For Curve Fit/14mA_clean.txt", "Data For Curve Fit/16mA_clean.txt", "Data For Curve Fit/18mA_clean.txt", 
        "Data For Curve Fit/20mA_clean.txt", "Data For Curve Fit/22mA_clean.txt", "Data For Curve Fit/24mA_clean.txt", 
        "Data For Curve Fit/26mA_clean.txt", "Data For Curve Fit/28mA_clean.txt"
]

for file in files:
    params, covariance = find_parameters(file)
    A, x0, sigma, offset = params
    print(f"File: {file}")
    print(f"A = {A}")
    print(f"x0 = {x0}")
    print(f"sigma = {sigma}")
    print(f"offset = {offset}")

    output_path = os.path.join(output_folder, "results.txt")
    with open(output_path, 'a') as f:
        f.write(f"File: {file}\n")
        f.write(f"A = {A}\n")
        f.write(f"x0 = {x0}\n")
        f.write(f"sigma = {sigma}\n")
        f.write(f"offset = {offset}\n\n")
        f.write("-" * 20 + "\n\n")
```
