To model each spectrum, I used a Gaussian function of the form:

\begin{equation}
I(\lambda) = A \exp\left(-\frac{(\lambda - x_0)^2}{2\sigma^2}\right) + \text{offset}
\end{equation}

where $A$ is the amplitude of the peak, $x_0$ is the peak wavelength, $\sigma$ represents the width of the peak, and the offset accounts for the baseline noise in the data. This model works well because the laser peaks are approximately Gaussian in shape.

For each dataset, I used the \texttt{curve\_fit} function in Python to find the values of $A$, $x_0$, $\sigma$, and the offset. The initial guesses for these parameters were taken directly from the data (for example, the peak intensity and its location). The fitting process minimizes the difference between the Gaussian model and the actual data.

After finding these parameters for each current, I treated them as functions of current:
\begin{equation}
A = A(I), \quad x_0 = x_0(I), \quad \sigma = \sigma(I), \quad \text{offset} = \text{offset}(I)
\end{equation}

Instead of forcing these into a specific equation, I used interpolation in Python (\texttt{interp1d}) so the model follows the actual data more closely. This avoids issues that came up when trying to fit everything to a straight line, which caused unrealistic peaks.

The final model combines everything into one equation:

\begin{equation}
I(\lambda, I) = A(I)\exp\left(-\frac{(\lambda - x_0(I))^2}{2\sigma(I)^2}\right) + \text{offset}(I)
\end{equation}

This allows me to input any current value and generate a predicted spectrum over a range of wavelengths.

To make sure the model stays physically reasonable, I added bounds in the code so that the amplitude and width stay positive, and the intensity cannot go negative. This was because at the lower currents, the intensity peak was going negative.

Overall, the method is: fit each dataset, extract the parameters, model how they change with current, and then use that to recreate the spectra.
